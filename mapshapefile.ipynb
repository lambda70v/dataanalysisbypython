```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline

fig, axes = plt.subplots()
shapefile_path = 'ch08/PortAuPrince_Roads/PortAuPrince_Roads'
m = Basemap(ax=axes,projection='stere', lon_0=-73, lat_0=18.75,
           llcrnrlat=17.25, urcrnrlat=20.25,
           llcrnrlon=-75, urcrnrlon=-71,
           resolution='f')
lines = m.readshapefile(shapefile_path, 'rivers')
m.drawrivers()
m.drawcountries()
m.drawcoastlines()
m.drawstates()

plt.show()
```
